```python
#Imports necessary packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loads each dataset as an np array
url = "https://raw.githubusercontent.com/kristabogan/neur265/main/spikes_right_trial1.csv"
url1 = "https://raw.githubusercontent.com/kristabogan/neur265/main/spikes_right_trial2.csv"
url2 = "https://raw.githubusercontent.com/kristabogan/neur265/main/spikes_left_trial1.csv"
url3 = "https://raw.githubusercontent.com/kristabogan/neur265/main/spikes_left_trial2.csv"

spikes_right_trial1 = np.loadtxt(url, delimiter=",")
spikes_right_trial2 = np.loadtxt(url1, delimiter=",")
spikes_left_trial1 = np.loadtxt(url2, delimiter=",")
spikes_left_trial2 = np.loadtxt(url3, delimiter=",")
```

```python
# Prints the length of each array
print(len(spikes_right_trial1))
print(len(spikes_right_trial2))
print(len(spikes_left_trial1))
print(len(spikes_left_trial2))
```

```python
# determines the firing rate of this neuron during each trial 
print((len(spikes_right_trial1))/30)
print((len(spikes_right_trial2))/30)
print((len(spikes_left_trial1))/30)
print((len(spikes_left_trial2))/30)
```

# Question 7
The highest firing rates occur on the second right trial and the first left trial. It is interesting that the first trial for the right turn has a lower firing rate, while the reverse is true for the left turn trials. I am unsure what this could mean- but maybe, the rat's brain is quickly adapting to the right turn trial.

# Question 9
This neuron looks like it fires more regularly during the left turn trials. Visually, it seems like the spikes are more evenly spread in the left turn trial than the right. In other words, the neuron during the right turn trial seems to have more quick-fire, successive bursts.

```python
#Calculates the ISIs for all four arrays
ISIsRight1 = np.diff(spikes_right_trial1)
ISIsLeft1 = np.diff(spikes_left_trial1)
ISIsRight2 = np.diff(spikes_right_trial2)
ISIsLeft2 = np.diff(spikes_left_trial2)
```

```python
#Plots the ISIs for all four arrays
bins = np.arange(0, 2, 0.01)
fig, ax = plt.subplots(2,2,figsize=(15,10))
ax[0,0].hist(ISIsRight1, bins)
ax[0,0].set_xlabel('Time [s]') 
ax[0,0].set_ylabel('ISI')
ax[0,0].set_title("ISI Values for the First Right Turn Trial")
ax[0,1].hist(ISIsRight2, bins)
ax[0,1].set_xlabel('Time [s]') 
ax[0,1].set_ylabel('ISI')
ax[0,1].set_title("ISI Values for the Second Right Turn Trial")
ax[1,0].hist(ISIsLeft1, bins)
ax[1,0].set_xlabel('Time [s]')
ax[1,0].set_ylabel('ISI')
ax[1,0].set_title("ISI Values for the First Left Turn Trial")
ax[1,1].hist(ISIsLeft2, bins)
ax[1,1].set_xlabel('Time [s]') 
ax[1,1].set_ylabel('ISI')
ax[1,1].set_title("ISI Values for the Second Left Turn Trial")
plt.show()
```

# Question 12

Yes! It seems like the neuron tends to fire in bursts. In particular, it fires in rapid succession at the beginning of each trial, with increasingly spaced out bursts as the trials go on. If I had to choose, I would say that the right trials have higher "burstiness" than in the left trial. In the left trials, it seems like single neuronal fires are more common. In other words, in the right trials, it seems like the neuron is more likely to fire in bursts.

```python
# Sets up the bins for all 4 subplots
binsize = 1
fr_bins = np.arange(0,30,binsize)                                        
fr_right1 = np.zeros(len(fr_bins)-1)                                                           
for i in range(len(fr_bins)-1):                                            
  spk_index = (spikes_right_trial1 > fr_bins[i]) & (spikes_right_trial1 < fr_bins[i+1])          
  true_values = spikes_right_trial1[spk_index]                                         
  fr_right1[i] = (len(true_values))/binsize

fr_bins = np.arange(0,30,binsize)                                        
fr_right2 = np.zeros(len(fr_bins)-1)                                                           
for i in range(len(fr_bins)-1):                                            
  spk_index = (spikes_right_trial2 > fr_bins[i]) & (spikes_right_trial2 < fr_bins[i+1])          
  true_values = spikes_right_trial2[spk_index]                                         
  fr_right2[i] = (len(true_values))/binsize

fr_bins = np.arange(0,30,binsize)                                        
fr_left1 = np.zeros(len(fr_bins)-1)                                                           
for i in range(len(fr_bins)-1):                                            
  spk_index = (spikes_left_trial1 > fr_bins[i]) & (spikes_left_trial1 < fr_bins[i+1])          
  true_values = spikes_left_trial1[spk_index]                                         
  fr_left1[i] = (len(true_values))/binsize

fr_bins = np.arange(0,30,binsize)                                        
fr_left2 = np.zeros(len(fr_bins)-1)                                                           
for i in range(len(fr_bins)-1):                                            
  spk_index = (spikes_left_trial2 > fr_bins[i]) & (spikes_left_trial2 < fr_bins[i+1])          
  true_values = spikes_left_trial2[spk_index]                                         
  fr_left2[i] = (len(true_values))/binsize
```

```python
# Plots the firing for all 4 trials in a 2 x 2 subplot structure, with right turns in blue, and left turns in red.
fig, ax = plt.subplots(2,2,figsize=(15,10))
ax[0,0].plot(fr_bins[:-1], fr_right1, "b")
ax[0,0].set_xlabel('Time [s]')
ax[0,0].set_ylabel('Firing Rate [Hz]')
ax[0,0].set_title("Firing Rate for the First Right Turn Trial")
ax[0,1].plot(fr_bins[:-1], fr_right2, "b")
ax[0,1].set_xlabel('Time [s]')
ax[0,1].set_ylabel('Firing Rate [Hz]')
ax[0,1].set_title("Firing Rate for the Second Right Turn Trial")
ax[1,0].plot(fr_bins[:-1], fr_left1, "r")
ax[1,0].set_xlabel('Time [s]')
ax[1,0].set_ylabel('Firing Rate [Hz]')
ax[1,0].set_title("Firing Rate for the First Left Turn Trial")
ax[1,1].plot(fr_bins[:-1], fr_left2, "r")
ax[1,1].set_xlabel('Time [s]')
ax[1,1].set_ylabel('Firing Rate [Hz]')
ax[1,1].set_title("Firing Rate for the Second Left Turn Trial")
```
